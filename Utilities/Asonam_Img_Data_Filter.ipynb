```python
import os
```

```python
data_dir = 'ASONAM17_Damage_Image_Dataset/'
data_files = ['gg.train', 'gg.test', 'gg.dev', 'nepal.train', 'nepal.test', 'nepal.dev','ecuador.train', 'ecuador.test', 'ecuador.dev']
```

```python
high_damage = []
low_damage = []
moderate_damage = []

filtered_path = 'filtered/asonam/'
high_dmg = 'high/'
low_dmg = 'low/'
moderate_dmg = 'moderate/'
```

```python
for df in data_files:
    file = open(data_dir + df, 'r') 
    for line in file:
        try:
            if ( int(line.replace('\n','').split(' ')[-1]) == 2):
                os.rename( data_dir + str(line.replace('\n','').split(' ')[0]) , filtered_path + high_dmg + str(line.replace('\n','').split(' ')[0].split('/')[-1]))

            elif ( int(line.replace('\n','').split(' ')[-1]) == 1):
                os.rename( data_dir + str(line.replace('\n','').split(' ')[0]) , filtered_path + moderate_dmg + str(line.replace('\n','').split(' ')[0].split('/')[-1]))

            else:
                os.rename( data_dir + str(line.replace('\n','').split(' ')[0]) , filtered_path + low_dmg + str(line.replace('\n','').split(' ')[0].split('/')[-1]))
        except Exception as e:
            print(e)
```
